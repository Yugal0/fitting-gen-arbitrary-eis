```python
import numpy as np
x = np.linspace(0, 1, 10)
y = 1000*np.exp(x)
```

```python
xnew = np.linspace(0, 1, num=1001)
ynew = np.interp(xnew, x, y)
ynew2 = np.interp(0.5, x, y)
```

```python
import matplotlib.pyplot as plt
plt.plot(xnew, ynew, '-', label='linear interp')
plt.plot(x, y, 'o', label='data')
plt.legend(loc='best')
plt.show()
```
